```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Matplotlib

```python
data = pd.read_csv('./DataSets/tips.csv')
data.head()
```

#### Scatter

```python
plt.scatter(data['day'], data['tip'])
plt.title('Tip by Day')
plt.xlabel('Day')
plt.ylabel('Tip')

plt.show()
```

#### Line chart

```python
plt.plot(data['tip'], label='tip')
plt.plot(data['size'], label='size')

plt.legend()

plt.show()
```

#### Bar chart

```python
plt.bar(data['day'], data['tip'])

plt.title('Bar chart')

plt.show()
```

#### Histogram

```python
plt.hist(data['total_bill'])

plt.title('Histogram chart')
plt.show()
```
