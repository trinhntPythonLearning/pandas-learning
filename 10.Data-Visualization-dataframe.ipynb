#### Pie chart

```python
irisData = pd.read_csv('./DataSets/iris.data')
irisData.columns = ["sepal_length","sepal_width", "petal_length", "petal_width", "species"]
irisData.head()
```

```python
iris_avg = irisData['petal_width'].groupby(irisData['species']).mean()
iris_avg
```

```python
iris_avg.plot.pie()
```

```python
iris_avg.plot.pie(labels=['Setosa', 'Versi', 'Virginica'], colors=list('bgr'), autopct='%.2f')
```
