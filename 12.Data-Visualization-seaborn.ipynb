Seaborn is a library mostly used for statistical plotting in Python  
It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_frame = pd.read_csv('./DataSets/tips.csv')
data_frame.head()
```

```python
sns.lineplot(x='sex', y='total_bill', data=data_frame)
plt.title('Bar Chart')
plt.show()
```

```python
sns.barplot(x='day', y='tip', data=data_frame, hue='sex')

plt.title('Bar Chart')
plt.show()
```

```python
sns.histplot(x='total_bill', data=data_frame, kde=True, hue='sex')

plt.title('History chart')
plt.show()
```

#### Data Analysis with Python using Forbes 2022 Dataset

```python
fb_df = pd.read_csv('./DataSets/forbes_2022_billionaires.csv')
fb_df.head()
```

```python
print(fb_df.shape)
print(fb_df.columns)
```

```python
fb_df['gender'].value_counts()
```

```python
gender_data = fb_df.groupby('gender')
gender_data.size()
```

```python
gender_data.size().plot.bar()
```

Who are the most richest?

```python
sns.barplot(y=fb_df['personName'][:10], x=fb_df['finalWorth'][:10])
```

Which country has highest number of richest? 

```python
country_df = fb_df.groupby('country').size().sort_values(ascending=False)
sns.barplot(y=country_df.index[:10], x=country_df.values[:10])
```

Top 10 richest in US

```python
fr_df = fb_df[fb_df['country']=='France']
fr_df.head()
```

```python
sns.barplot(y=fr_df['personName'][:10], x=fr_df['finalWorth'][:10])
plt.title('Top 10 Frech richest')
```

```python
sns.histplot(x=fb_df['age'])
```
