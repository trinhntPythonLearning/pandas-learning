Seaborn is a library mostly used for statistical plotting in Python  
It is built on top of Matplotlib and provides beautiful default styles and color palettes to make statistical plots more attractive

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_frame = pd.read_csv('./DataSets/tips.csv')
data_frame.head()
```

```python
sns.lineplot(x='sex', y='total_bill', data=data_frame)
plt.title('Bar Chart')
plt.show()
```

```python
sns.barplot(x='day', y='tip', data=data_frame, hue='sex')

plt.title('Bar Chart')
plt.show()
```

```python
sns.histplot(x='total_bill', data=data_frame, kde=True, hue='sex')

plt.title('History chart')
plt.show()
```
