```python
import pandas as pd
import numpy as np
```

```python
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
    
  
# Convert the dictionary into DataFrame 
df = pd.DataFrame(data1)
  
df
```

```python
# split data to group
grps = df.groupby('Name')
print(type(grps))
print()

# Iterating group
for name, grp in grps:
    print(name)
    print(grp)
    print()
```

```python
# Get group
grps.get_group('Jai')
```

```python
# Aggregate Data

print(grps.aggregate(np.sum))
print()
print(grps['Age'].agg([np.sum, np.mean, np.std]))
```

```python
# Filter data

df
```

```python
grps = df.groupby('Name')

filterData = grps.filter(lambda x: len(x) >= 2)
filterData
```

Combine multiple column groupby with dictionary

```python
d = {'id':['1', '2', '3'],
     'Column 1.1':[14, 15, 16],
     'Column 1.2':[10, 10, 10],
     'Column 1.3':[1, 4, 5],
     'Column 2.1':[1, 2, 3],
     'Column 2.2':[10, 10, 10], }
  
# Converting dictionary into a data-frame 
df = pd.DataFrame(d)
df = df.set_index('id')
df
```

```python
groupby_dict = {'Column 1.1':'Column 1',
                'Column 1.2':'Column 1',
                'Column 1.3':'Column 1',
                'Column 2.1':'Column 2',
                'Column 2.2':'Column 2' }

min_df = df.groupby(groupby_dict, axis=1).min()
sum_df = df.groupby(groupby_dict, axis=1).sum()
print(min_df)
print()
print(sum_df)
```
